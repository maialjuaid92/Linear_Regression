```python
#using BeautifulSoup to read the webpage as HTML
page = requests.get(weather_Link).text
soup = BeautifulSoup(page, 'html5lib')
```

```python
driver = webdriver.Chrome(chromedriver)
driver.get(weather_Link)
```

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
#locating the table
contents_tbody=soup.find('tbody')
```

```python
#displaying the table contents (HTMLcode)
contents_tbody
```

```python
 #locating table columns by class name
soup.find('td', class_ = "ng-star-inserted")
```

```python
#looping over each column in the table to fetch data from the weather webpage
data_time = []
table = soup.find('table', attrs={'class':'days'})
table_body = table.find('tbody')


rows = table_body.find_all('tr')

for row in rows:
    cols = row.find_all('td')
    cols8 = [ele.text.strip() for ele in cols]
    data_time.append([ele for ele in cols8 if ele]) # Get rid of empty values
    
    
    
    
```

```python
#looping over each day in the table
f_MAx=[]
for ddd in data_time:
    if 'Max' not in ddd:
        f_MAx.append(ddd)
```

```python
#creating dataframes for each column by specifying index

df1=pd.DataFrame(f_MAx[1:32])
df2=pd.DataFrame(f_MAx[32:63])
df3=pd.DataFrame(f_MAx[63:94])
df4=pd.DataFrame(f_MAx[94:125])
df5=pd.DataFrame(f_MAx[125:156])
df6=pd.DataFrame(f_MAx[156:187])
df7=pd.DataFrame(f_MAx[188:235])
```
